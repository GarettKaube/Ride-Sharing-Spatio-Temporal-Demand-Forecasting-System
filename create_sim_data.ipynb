### Just a notebook which was used to generate the simulation data located in src/data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import duckdb
import numpy as np
```

```python
conn = duckdb.connect("rides.ddb")
```

```python
query = """
    SELECT * FROM RIDES_EXTRA_VARS
    LIMIT 10000
"""
df = conn.execute(query).fetch_df()
```

```python
df
```

```python
query = """
    WITH communities AS (
        SELECT *,
        ROW_NUMBER() OVER(
            PARTITION BY pickup_community_area 
            ORDER BY trip_start_timestamp DESC
        ) AS row_number_for_community
        FROM RIDES_EXTRA_VARS 
        WHERE pickup_community_area IN (32, 8, 33, 28, 31, 24)
    )

    SELECT *
    FROM communities
    WHERE row_number_for_community <= 50000
"""

df = conn.execute(query).fetch_df()
```

```python
df = df.sort_values(by = "trip_start_timestamp")
```

```python
df.to_csv("simulation_data_communities.csv", index=None)
```
